```python
#board
#display board
#give input
#check win
    #check row
    #check column
    #check diagonal
#restart game
def play_again():
    board = ["-","-","-",
            "-","-","-",
            "-","-","-"]
    return board
def display_board(board):
    print(board[0]+" | "+board[1]+" | "+board[2]) 
    print(board[3]+" | "+board[4]+" | "+board[5])
    print(board[6]+" | "+board[7]+" | "+board[8])



def driver_function():
    board=play_again()
    name1=input("enter the name of player 1:")
    name2=input("enter the name of player 2:")
    if_win=False
    c=0
    while if_win==False and c<9:
        board=take_input('X',name1,board)
        if_win=check_win(name1,name2,board)
        c+=1
        if if_win==False and c<9:
            board=take_input('O',name2,board)
            if_win=check_win(name1,name2,board)
            c+=1
        else:
            break
    display_board(board)
    if c==9 and if_win==False:
        print("The match ended in a tie.")
    restart=input("Do you want to play again?")
    if restart=='yes':
        driver_function()
    else:
        print("Thanks for playing.")
    

def take_input(symbol, name, board):
    display_board(board)
    inp=int(input(name+", enter the position to place "+symbol))
    board[inp-1]=symbol
    return board

    
def check_win(name1, name2, board):
    win=""
    win=check_row(board)
    if win=="X":
        print(name1+", congratulations you are the winner!!")
        return True
    elif win=="O":
        print(name2+", congratulations you are the winner!!")
        return True
    win= check_column(board)
    if win=="X":
        print(name1+", congratulations you are the winner!!")
        return True
    elif win=="O":
        print(name2+", congratulations you are the winner!!")
        return True
    win=check_diagonal(board)
    if win=="X":
        print(name1+", congratulations you are the winner!!")
        return True
    elif win=="O":
        print(name2+", congratulations you are the winner!!")
        return True
    return False
def check_row(board):
    row1=False
    row2=False
    row3=False
    if board[0]==board[1]==board[2]!='-':
        row1=True
    elif board[3]==board[4]==board[5]!='-':
        row2=True
    elif board[6]==board[7]==board[8]!='-':
        row3=True
    if row1:
        return board[0]
    elif row2:
        return board[3]
    elif row3:
        return board[6]
def check_column(board):
    col1=False
    col2=False
    col3=False
    if board[0]==board[3]==board[6]!='-':
        col1=True
    elif board[1]==board[4]==board[7]!='-':
        col2=True
    elif board[2]==board[5]==board[8]!='-':
        col3=True
    if col1:
        return board[0]
    elif col2:
        return board[1]
    elif col3:
        return board[2]

def check_diagonal(board):
    dia1=False
    dia2=False
    if board[0]==board[4]==board[8]!='-':
        dia1=True
    elif board[2]==board[4]==board[6]!='-':
        dia2=True
    if dia1:
        return board[0]
    elif dia2:
        return board[2]
    
driver_function()
```
